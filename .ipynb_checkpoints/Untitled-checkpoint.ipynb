```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
```

```python
(train_data , train_label ) , (test_data , test_label) = mnist.load_data()
```

```python
train_data = train_data / 255
test_data  = test_data / 255
```

```python
print("Shape of train  data : " , train_data.shape)
print("Shape of train label : " , train_label.shape)
print("Shape of test   data : " , test_data.shape)
print("Shape of test  label : " , test_label.shape)
```

```python
train_label = tf.keras.utils.to_categorical(train_label)
test_label  = tf.keras.utils.to_categorical(test_label)
```

```python
print("Shape of train  data : " , train_data.shape)
print("Shape of train label : " , train_label.shape)
print("Shape of test   data : " , test_data.shape)
print("Shape of test  label : " , test_label.shape)
```

```python
plt.imshow(train_data[0] , cmap = plt.cm.binary_r)
plt.title("Digit : %s"%str(train_label[0].argmax()))
```

```python
X_train = train_data [ : 50000 , : , : ]
y_train = train_label[ : 50000 , : ]
X_valid = train_data [50000 : , : , : ]
y_valid = train_label[50000 : , : ]
```

```python
print("Shape of X_train : " , X_train.shape)
print("Shape of y_train : " , y_train.shape)
print("Shape of X_valid : " , X_valid.shape)
print("Shape of y_valid : " , y_valid.shape)
```

```python
def model(layer , l2 = 0 , dropout = 0):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Flatten(input_shape = (28 , 28)))
    for l in layer:
        model.add( tf.keras.layers.Dense(l , kernel_regularizer = tf.keras.regularizers.l2(0.0001) , activation = "relu" ) )
        model.add( tf.keras.layers.Dropout(dropout) )
    model.add( tf.keras.layers.Dense( 10 , activation = "softmax" ) )
    model.compile(optimizer = "rmsprop" , loss = tf.keras.losses.categorical_crossentropy , metrics = ['acc'])
    return model
```

```python
model1 = model([128 , 128])
his1 = model1.fit(X_train , y_train ,
                  epochs = 30 ,
                  verbose = 0 ,
                  batch_size = 128 ,
                  validation_data = (X_valid , y_valid))
```

```python
def plot_history(history):
    _ , ax = plt.subplots(1 , 2 , figsize = (18,6))

    ax[0].plot(history.history['acc'] , label = "training")
    ax[0].plot(history.history['val_acc'] , label = "validation")
    ax[0].set_title("Accurecy")
    ax[0].legend()

    ax[1].plot(history.history['loss'] , label = "training")
    ax[1].plot(history.history['val_loss'] , label = "validation")
    ax[1].set_title("Croosentropy")
    ax[1].legend()
```

```python
plot_history(his1)
```

```python
model2 = model([128 , 128] , dropout = 0.5)
his2 = model2.fit(X_train , y_train ,
                  epochs = 30 ,
                  verbose = 0 ,
                  batch_size = 128 ,
                  validation_data = (X_valid , y_valid))
plot_history(his2)
```
