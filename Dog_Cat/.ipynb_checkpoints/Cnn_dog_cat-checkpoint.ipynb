```python
import tensorflow as tf
print("Version of tensorflow is : " , tf.__version__)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.applications import VGG16
```

```python
import os
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("./train/dogs/dog.10.jpg")
img2 = cv2.imread("./train/cats/cat.0.jpg")
```

```python
plt.imshow(img1)
```

```python
train_dir = "./train/"
test_dir  = "./validation/"
```

```python
print("Train Details :\n===========================")
for i in os.listdir(train_dir):
    print("No of images in "+i+" is : ",len(os.listdir(train_dir+i)))
print("Validation Details :\n===========================")
for i in os.listdir(test_dir):
    print("No of images in "+i+" is : ",len(os.listdir(test_dir+i)))
```

```python
train_datagen = ImageDataGenerator(rescale = (1.0 / 255.0))
test_datagen  = ImageDataGenerator(rescale = (1.0 / 255.0))
```

```python
train_data = train_datagen.flow_from_directory(train_dir ,
                                               target_size = (150 , 150) ,
                                               batch_size = 20 ,
                                               class_mode = "binary")
test_data  = test_datagen.flow_from_directory(test_dir ,
                                              target_size = (150 , 150) ,
                                              batch_size = 20 ,
                                              class_mode = "binary")
```

# CNN for dog cat

```python
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32 , (3,3) , activation = "relu" , input_shape = (150 , 150 , 3)),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Conv2D(64 , (3,3) , activation = "relu"),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Conv2D(128 , (3,3) , activation = "relu"),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Conv2D(128 , (3,3) , activation = "relu"),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512 , activation = "relu"),
    tf.keras.layers.Dense(1 , activation = "sigmoid")
])
model1.summary()
```

```python
model1.compile(optimizer = "rmsprop" , loss = tf.keras.losses.binary_crossentropy , metrics = ['acc'])
```

```python
train_data.n
```

```python
train_data.class_indices
```

```python
his1 = model1.fit(train_data ,
                  steps_per_epoch = train_data.n // train_data.batch_size ,
                  epochs = 25 ,
                  validation_data = test_data ,
                  validation_steps = test_data.n // test_data.batch_size)
```

```python
model1.save("overfittedmodel.h5")
```

```python
import seaborn as sns
sns.set_style("darkgrid")
def plot_history(history):
    _ , ax = plt.subplots(1 , 2 , figsize = (18,6))

    ax[0].plot(history.history['acc'] , label = "training")
    ax[0].plot(history.history['val_acc'] , label = "validation")
    ax[0].set_title("Accurecy")
    ax[0].legend()

    ax[1].plot(history.history['loss'] , label = "training")
    ax[1].plot(history.history['val_loss'] , label = "validation")
    ax[1].set_title("Croosentropy")
    ax[1].legend()
```

```python
plot_history(his1)
```

# Data Augmentatin

```python
train_data_aug = ImageDataGenerator(rescale = (1.0 / 255.0) ,
                                    rotation_range = 40 ,
                                    height_shift_range = 0.2 ,
                                    width_shift_range = 0.2 ,
                                    shear_range = 0.2 ,
                                    zoom_range = 0.2 ,
                                    fill_mode = "nearest")
test_data_aug  = ImageDataGenerator(rescale = (1.0 / 255.0))
```

```python
train_data_aug = train_data_aug.flow_from_directory(train_dir ,
                                                    batch_size = 20 ,
                                                    target_size = (150 , 150) ,
                                                    class_mode = "binary")

test_data_aug = test_data_aug.flow_from_directory(test_dir ,
                                                  batch_size = 20 ,
                                                  target_size = (150 , 150) ,
                                                  class_mode = "binary")
```

# Model

```python
model2 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32 , (3,3) , activation = "relu" , input_shape = (150 , 150 , 3)),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Conv2D(64 , (3,3) , activation = "relu"),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Conv2D(128 , (3,3) , activation = "relu"),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Conv2D(128 , (3,3) , activation = "relu"),
    tf.keras.layers.MaxPool2D((2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512 , activation = "relu"),
    tf.keras.layers.Dense(1 , activation = "sigmoid")
])

model2.compile(optimizer = "rmsprop" , loss = tf.keras.losses.binary_crossentropy , metrics = ['acc'])

model2.summary()
```

```python
his2 = model2.fit(train_data_aug ,
                  steps_per_epoch = train_data_aug.n // train_data_aug.batch_size ,
                  epochs = 25 ,
                  validation_data = test_data_aug ,
                  validation_steps = test_data_aug.n // test_data_aug.batch_size)
```

```python
plot_history(his2)
```

```python
model2.save("augmentedmodel.h5")
```

## Pretrained Mode - VGG16

```python
conv_base = VGG16(weights = "imagenet" ,
                  include_top = False ,
                  input_shape = (150 , 150 , 3))
```

```python
conv_base.summary()
```

```python
conv_base.trainable = False
```

```python
model = tf.keras.models.Sequential()
model.add(conv_base)
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(256 , activation = "relu"))
model.add(tf.keras.layers.Dense(1 , activation = "sigmoid"))
```

```python
model.summary()
```

```python
train_data_aug = ImageDataGenerator(rescale = (1.0 / 255.0) ,
                                    rotation_range = 40 ,
                                    height_shift_range = 0.2 ,
                                    width_shift_range = 0.2 ,
                                    shear_range = 0.2 ,
                                    zoom_range = 0.2 ,
                                    horizontal_flip = True ,
                                    fill_mode = "nearest")

test_data_aug  = ImageDataGenerator(rescale = (1.0 / 255.0) )

train_data_aug = train_data_aug.flow_from_directory(train_dir ,
                                                    batch_size = 20 ,
                                                    target_size = (150 , 150) ,
                                                    class_mode = "binary")

test_data_aug = test_data_aug.flow_from_directory(test_dir ,
                                                  batch_size = 20 ,
                                                  target_size = (150 , 150) ,
                                                  class_mode = "binary")
```

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self , epoch , logs = {}):
        if logs.get('acc')>0.95:
            print("\nReached 95.0% accuracy so cancelling training!")
            self.model.stop_training = True
```

```python
callbacks = myCallback
his = model.fit(train_data_aug ,
                  steps_per_epoch = train_data_aug.n // train_data_aug.batch_size ,
                  epochs = 25 ,
                  validation_data = test_data_aug ,
                  validation_steps = test_data_aug.n // test_data_aug.batch_size)
```
